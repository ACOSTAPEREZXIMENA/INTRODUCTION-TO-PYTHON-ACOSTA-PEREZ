## If-else function:
Its function is to control the program and allow you to execute code blocks according to the given values, if the condition of the if statement is fulfilled, it will be executed correctly, otherwise the else statement will be fulfilled. ## Examples

```python
age= int(input("write your age"))
if age >60:
 print ("You are an elderly person")
else:
 print("You are a young adult")
```

This is an example giving as a situation to know if you belong to the young adults or to the adults of the third age, this code I elaborated it first making a variable and placing the if, saying if it is more than 60 it is older adult and if it is smaller then it is young adult. 

```python
unNumeroFloat = .40
alcoholFloat = float(input("write the percentage of alcohol obtained: "))
if alcoholFloat >= unNumeroFloat:
 print("Exceeded consumption of alcohol, must pay fine")
else:
 print("I pass the Breathalyzer , you may continue")
```

The situation presented is a breathalyzer, that is if you are over .40 you were already over the limit and would have to pay a fine, for this code I base it on if it is over .40 then you pay and if it is under you pass free, that is why I used if else. 

```python
people = int(input("Please enter your age to find out which dish you will be matched with."))
if people >=12:
 print("The saucer will be adult")
else:
 print("The saucer will be a child's saucer")
```

## Function of the while Loop:
In python a loop is characterized by the fact that thanks to it we can repeat the instructions or the body of the loop as many times as we want as long as the condition is met which is that it is true, in case it is false it can not be executed as required, that is why the condition must be met.[Tutorial de paython](https://tutorial.recursospython.com/bucles/)(2013)



```python
i = 20
while i > 0:
  print(i)
  i -= 2
```

This example is an inverse counter that goes from 2 to 2 and I made it by putting a while and the indication of how I wanted it to be. 


```python
Z = 4
while Z > 0:
    Z-=1
    print(Z)
```

In this example we have the situation of a condition, and d a block as long as the condition is met.

```python
a=0
while a<10:
    a+=1
    if a==8:
        continue
    print(a)
```

Finally we have this one which is a cycle that continues counting, however when it reaches the set number it stops and cuts off.

## Function of the for loop:
A for loop has the purpose of repeating blocks of instructions or list of data, giving them an assigned or predetermined value
[Franciscomelov](https://www.freecodecamp.org/espanol/news/bucles-for-en-python/)(2021)

```python
print("Inicio")
for i in [0, 1, 2]:
    print("Hi ", end="")
print()
print("Finish")
```

Is the one who is in charge of the step-by-step progress of the program, he is the one who is in control.

```python
for q in range(100, -1, -1):
    print(q)
```

It is a for-cycle that performs its procedure in reverse, i.e., decreasing the step value, from 100 to 0.

```python
for j in range(4):
    print(j+1)
```

This is a counter that goes one at a time and I only made it by putting a range and a variable in it.
